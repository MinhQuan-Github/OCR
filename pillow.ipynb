# Padding

```python
right = 98
left = 105
top = 100
bottom = 33
```

```python
width, height = img.size
```

```python
new_width = width + right + left
new_height = height + top + bottom
```

```python
padding_img = Image.new(img.mode, (new_width, new_height), (0, 0, 255))
padding_img.paste(img, (left, top))
```

```python
padding_img
```
